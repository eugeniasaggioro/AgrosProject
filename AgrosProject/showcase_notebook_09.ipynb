### Correlation between variables

The World's agricultural production dataset offers information about inputs and outputs that may influence a country's total production. By examining the quantities relative to inputs and outputs, it becomes clear that these variables are positively and strongly correlated among themselves. 

To be more precise, all of the variables have a minimum correlation coefficient of 0.26 and the majority of them are correlated with a coefficient higher than 0.7. This shows that rising levels of one quantitative variable are typically accompanied by rising levels of others.
Crop output and total output are perfectly correlated among the output quantities, which is not surprising given that crops are most likely the major contributor to agricultural output. 
Because that farm machinery is frequently used to cultivate and harvest crops, it makes natural that farm machinery, cropland, and fertilizer have a strong correlation.

However, aside from the amount of machinery and crop output, fish output has the lowest correlation with the other variables, in particular with pasture, cropland and agricultural land quantity. This might be due to the fact that, in comparison to crops and livestock, aquaculture makes up a very tiny portion of agricultural production overall.

Out of all the variables, pasture quantity has the lowest correlation with the others. Pasture quantity refers to the amount of land used for grazing livestock, which is mainly used for producing meat, milk, or other animal products. While it is an essential aspect of many agricultural systems, it may have different impacts and uses compared to other agricultural practices such as crop cultivation or greenhouse farming. Thus, the amount of pasture used may not always directly correlate with the quantity of fertilizer used or the amount of crops produced.

The amount of animal output is strongly correlated with the Mcal of metabolizable energy (animal_feed_quantity) in animal feed and also with the amount of capital and fertilizer. This shows that ensuring animals have a proper diet is essential for raising animal productivity. Labor quantity has a higher correlation with the amount of irrigation and livestock quantity and a lower correlation with fish, animal outputs, capital, and machinery among others. This suggests the areas where labor is more needed. 

```python
Agros.plot_quantity_correlations()
```

The strong positive correlations observed among almost all variables suggest that increasing any single input or output quantity has the potential to drive overall productivity gains. Our analysis reveals that, over time, the overall output has consistently increased in the countries we analyzed, and this growth is consistently associated with increases in input quantities. Essentially, countries with greater input potential are also better positioned to achieve higher output levels.

### TFP: How the Total Factor Productivity changed in the last 20 years

The choropleth map presented below provides a clear visual representation of the global distribution of the relationship between agricultural outputs and inputs.  It facilitates comprehension and insight into how TFP varies across different geographical locations.

Let's see how the TFP changed from 2000 to 2019.

```python
Agros.choropleth(2000)
```

```python
Agros.choropleth(2019)
```

By keeping the scale of color with the same range of values in both graphs we can state that in general the map corresponding to 2019 has lighter colors compare to the one of 2000. This, means that the TFP increased in the last 20 years, almost around the whole world.

A higher value of TFP indicates greater efficiency and productivity, meaning that the economy, industry, or firm is producing more output with the same level of inputs or producing the same output with fewer inputs. On the other hand, a lower value of TFP would suggest that the economy, industry, or firm is less efficient and productive.

It appears that there is a decrease in Total Factor Productivity (TFP) in only a small number of African countries based on the available data. This could suggest that TFP trends in Africa are relatively stable, with only a few exceptions where productivity is declining. However, it's important to note that this conclusion is based on limited information and may not reflect the full picture.
